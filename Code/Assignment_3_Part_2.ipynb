## **üõ≥Ô∏è Titanic Dataset**
The Titanic dataset is a collection of data pertaining to the passengers aboard the RMS Titanic, which famously sank on its maiden voyage on April 15, 1912. The task is to conduct Exploratory Data Analysis(EDA) and investigate minimum three key hypotheses related to passenger class, gender, and age in relation to survival rates. In doing so, we aim to uncover patterns and insights within this iconic dataset. 


### üëâ Data fields
* PassengerId - a unique identifier for each passenger
* Survived - indicates whether the passenger survived (1) or did not survive (0)
* Pclass - passenger class (1st, 2nd, or 3rd)
* Name - the name of the passenger
* Sex - the gender of the passenger
* Age - the age of the passenger
* SibSp - the number of siblings or spouses aboard
* Parch - the number of parents or children aboard
* Ticket - the ticket number
* Fare - the fare paid for the ticket
* Cabin - the cabin number
* Embarked - the port where the passenger boarded (C = Cherbourg, Q = Queenstown, S = Southampton)

#### 2.3 Survival rate association with age

```python
sns.boxplot(data=df, y="Survived", x="Age", orient="h")
plt.title('Box Plot of Age by Survied')
```
